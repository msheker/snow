```python
import pandas as pd
import numpy as np
import numpy.linalg as la

import matplotlib.pyplot as plt
from bokeh.plotting import figure, show, output_file
from bokeh.layouts import gridplot, Row
from bokeh.io import hplot

data_all = pd.read_csv("gauge.csv")
data_all
```

```python
data = data_all

(inter, slope) = np.polyfit(data["density"], data["gain"],1)
z = (inter, slope)
p = np.poly1d(z)
pnts = [p(x) for x in np.unique(data["density"])]

res = []
sm = 0
for i in range(len(data["density"])):
    sm += abs(data["gain"][i] - p(data["density"][i]))
    res.append(sm)

#least fit, no transforms
f0 = figure(title = "Gain vs Density")
f0.scatter(data["density"], data["gain"])
f0.line(list(np.unique(data["density"])),pnts)
f0.line(list(np.unique(data["density"])),res, color = "red")

show(f0)
```

```python
data = data_all
data["log_gain"] = np.log(data["gain"])

(inter, slope) = np.polyfit(data["density"], data["log_gain"],1)
z = (inter, slope)
p = np.poly1d(z)
pnts = [p(x) for x in np.unique(data["density"])]

res = []
sm = 0
for i in range(len(data["density"])):
    sm += abs(data["log_gain"][i] - p(data["density"][i]))
    res.append(sm)
    
#least fit, log transform
f1 = figure(title = "log(Gain) vs Density")
f1.scatter(data["density"], data["log_gain"])
f1.line(list(np.unique(data["density"])),pnts)
f1.line(list(np.unique(data["density"])),res, color = "red")

show(f1)
```

```python
show(gridplot(f0,f1, ncols=2, plot_width=400, plot_height=400, toolbar_location=None))
```
